```python
import cv2
import numpy as np
import pydicom
   
from Tkinter import *
import Tkinter as tk
import Image, ImageTk, ImageDraw
import numpy as np
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import tkFileDialog
import tkMessageBox
 
cropping = False
 
x_start, y_start, x_end, y_end = 0, 0, 0, 0
root = tk.Tk()
root.withdraw()
tkMessageBox.showinfo("DCM-Crop video maker",'Select a DICOM file to crop')
#root.withdraw()
filename = tkFileDialog.askopenfilename()
ds = pydicom.read_file(filename)
ID=ds.PatientID
tkMessageBox.showinfo("Patient loaded:he ID of the patient loaded is", ID) 
root.destroy()
print ID


nf=ds.NumberOfFrames #is a string!
int(nf) #casted to integer
arr=ds.pixel_array#filename=ds.pixel_arra
 
image =arr[0]#

oriImage = image.copy()

 
def mouse_crop(event, x, y, flags, param):
    # grab references to the global variables
    global x_start, y_start, x_end, y_end, cropping,refPoint,cropped_frames,size,crop_imgs
 
    # if the left mouse button was DOWN, start RECORDING
    # (x, y) coordinates and indicate that cropping is being
    if event == cv2.EVENT_LBUTTONDOWN:
        x_start, y_start, x_end, y_end = x, y, x, y
        cropping = True
 
    # Mouse is Moving
    elif event == cv2.EVENT_MOUSEMOVE:#keep recording coordinates 
        if cropping == True:
            x_end, y_end = x, y
 
    # if the left mouse button was released
    elif event == cv2.EVENT_LBUTTONUP:
        # record the ending (x, y) coordinates
        x_end, y_end = x, y
        cropping = False # cropping is finished
 
        refPoint = [(x_start, y_start), (x_end, y_end)]
      
        cropped_frames=[]
        for i in range(len(arr)):
            crop=arr[i] [refPoint[0][1]:refPoint[1][1], refPoint[0][0]:refPoint[1][0]]
            cropped_frames.append(crop)
        if len(refPoint) == 2: #when two points were found
            roi = oriImage[refPoint[0][1]:refPoint[1][1], refPoint[0][0]:refPoint[1][0]]
            cv2.imshow("Cropped", roi)
            root = tk.Tk()
            root.withdraw()
            tkMessageBox.showinfo("Well Done!", "Frame has been cropped!...push Esc to exit")  
            root.destroy()
            #crop_imgs5 = arr[5][y_start:y_end, x_start:x_end]
            #crop_imgs80 = arr[80][y_start:y_end, x_start:x_end]
            #cv2.imshow("5 frame",crop_imgs5)
            #cv2.imshow("80 frame",crop_imgs80)
            #crops_frames=[]
       # for index in range (len(arr)):
            #crop_imgs = [arr[index][y_start:y_end, x_start:x_end]]#single frame in index of arr with cropped coordinates
            #crops_frames.append(crop_imgs)#total list with all frames array
           # cv2.imshow("5 frame",crops_frames[crop_imgs[5]])#try to print all frames)
cv2.namedWindow("image")#create a window to place amn image
cv2.setMouseCallback("image", mouse_crop)#specify the window on wich use the mouse_crop function

while True:
 
    copy = image.copy()
    k = cv2.waitKey(1) & 0xFF
 
    if not cropping:
        cv2.imshow("image", image)
 
    elif cropping:
        cv2.rectangle(copy, (x_start, y_start), (x_end, y_end), (255, 0, 0), 2)
        cv2.imshow("image", copy)
        
    if k == 27: 
        
        cv2.destroyAllWindows()# Esc key to stop
        
        break

#print (len(crop_imgs))#=1 no is the single cropped frames
#print(len(crops_frames)) #= 377 OK
#print(crop_imgs)
#print(arr[0])#is an array
#print(crops_frames[0])#is an array of array
#cv2.imshow("5 frame",crops_frames[[0][5]])
###nel video se stampo arr[5 lo fa se faccio arr[5][y_start:y_end, x_start:x_end] No
#print(len(arr[5]))#=600
#print(len(arr[5][y_start:y_end, x_start:x_end]))#= less than 600 (it depends on cropping)
#print(len(arr[80][y_start:y_end, x_start:x_end]))#its equal to the above one
#print(arr[5][y_start:y_end, x_start:x_end]).shape
def convert_frames_to_video(pathOut,fps):
   
    
 
    
 
   
        
      
        frames,height, width  = np.array(cropped_frames).shape
        size = (width,height)
        
  
 
        out = cv2.VideoWriter(pathOut,cv2.VideoWriter_fourcc(*'PIM1'), fps, size,False)
 
        for i in range(len(np.array(cropped_frames))):
        
            out.write(np.array(cropped_frames[i]))
        out.release()
 

  
 

pathOut = ds.PatientID + '.avi'
fps = 25.0#377 frames/25fps-->15 sec video(15 fps)
convert_frames_to_video( pathOut, fps)
```
